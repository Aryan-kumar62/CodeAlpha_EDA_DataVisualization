```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
pd.options.display.max_columns=None
pd.options.display.max_rows=None
```

```python
df=pd.read_csv("superstore.csv",encoding='latin-1')
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Category", y="Sales", data=df)
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

### Sales by Category
Technology category records the highest sales,
followed by Office Supplies and Furniture.


```python
plt.figure(figsize=(8,5))
sns.barplot(x="Region", y="Profit", data=df)
plt.title("Profit by Region")
plt.xlabel("Region")
plt.ylabel("Total Profit")
plt.show()
```

### Profit by Region
The West region is the most profitable,
while the Central region shows lower profit margins.


```python
plt.figure(figsize=(8,5))
sns.histplot(df["Sales"], bins=30)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

### Sales Distribution
Most sales values are concentrated at lower ranges,
with a few high-value orders.
